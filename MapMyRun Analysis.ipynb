Some dates came in a different format, which Excel didn't recognize as dates, therefore I had to convert these strings into a usable format. After completing this for all 334 entries I was still running into an issue, the built in date function did not like the month being input as the alphabetical abbreviation (Jan, Feb.. etc.). To solve this I embedded a vlookup within the cell that link to a table that had the numerical equivalent for each month. 

Next there were 5 workouts with 0 for distance, all 5 were indoor runs which did still have calories. To estimate distance for these took the total calories for the other 329 entries and divided by the total distance to get my calories per mile.

I've updated these, highlighting any rows that have been manually altered and bolding the cells with updated values. For calories I've used the average cal per mile times distance. 

Then to identify outliers I've sorted by % difference from usual calories per mile and % difference from usual time per mile. Many of them are explained by the notes, for example running with my dog or my girlfriend, or a couple that were just walks.

```python
def count(data):
    count = {}
    for entry in list(data):
        if entry in count:
            count[entry] += 1
        else:
            count[entry] = 1    
    return count
```
