```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
NBApotw = pd.read_csv('NBA_player_of_the_week.csv')
NBApotw
```

```python
NBApotw.apply(lambda x: x.count(), axis=0)
```

```python
#determines the total amount of weeks won
def count_weeks_won(data):
    count = {}
    for entry in list(data):
        if entry in count:
            count[entry] += 1
        else:
            count[entry] = 1    
    return count
```

```python
#returns the key with the highest value
def max_val_key(data):
    values = list(data.values())
    keys = list(data.keys())
    return keys[values.index(max(values))]
```

```python
player_tally = count_weeks_won(NBApotw['Player'])
team_tally = count_weeks_won(NBApotw['Team'])
```

```python
most_awarded_player = max_val_key(player_tally)
most_awarded_team = max_val_key(team_tally)
```

```python
print("The most awarded player is " + most_awarded_player + " with " + str(player_tally[most_awarded_player]) + " and the most awared team is the " + most_awarded_team + " with " + str(team_tally[most_awarded_team]))
```

```python
plt.hist(player_tally.values(), bins = 13)
plt.title('Number of Times POTW Awarded to Each POTW Winner')
plt.xlabel('Number of Times POTW')
plt.ylabel("Number of Players")
```

```python
plt.scatter(player_tally.keys(), player_tally.values())
plt.title('Player of the week awards for a team')
plt.ylabel("Weeks won")
```

```python
plt.boxplot(player_tally.values())
plt.title('Number of Times POTW Awarded to Each POTW Winner')
plt.ylabel("Number of Times POTW")
```

This boxplot and scatterplot tells me that a vast majority of playes won the award less than 5 times, as a matter of fact just under 250 of the 274 weeks evaluated were won by players who won the award 5 times or less. The box plot shows that there is one dramatic outlier (Lebron James) that has won the award almost twice as much as any other player. The boxplot is probably the better visualization in this case.

```python
def percent_won(data):
    percent_won = {}
    for item in data:
        percent_won[item] = data[item]/sum(data.values())
    return percent_won
```

```python
team_tally_percent = percent_won(team_tally)
```

```python
plt.pie(team_tally_percent.values())
plt.title("POTW Awards Won by Team")
labels = team_tally.keys()
plt.axis('equal')
```
